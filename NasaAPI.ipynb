```python
def fetch_space_data():
  try:
    response = requests.get(URL)
    if response.status_code == 200:
      data_list = response.json()

      for item in data_list:
        title = item.get("title")
        image_url = item.get("url")
        explanation = item.get("explanation")
        media_type = item.get("media_type")

        print("Hello Space Wanderer")
        print("-" * 40)
        print(f"Today our Destination: {title}")
        print("-"*40)
        if item.get(media_type) == "image":
          display(Image(url=image_url, width=600))
        else:
          print(f"Today is a vedio! Watch it here : {image_url}")
        print("-" * 40)
        print(f"Did you know?: {explanation[:150]}...")
    else:
      print("Connection failed")
  except Exception as e:
    print(f"System Faliure: {e}")

fetch_space_data()
```

```python
def launch_time_machine():
  print("Welcome to nasa api time machine")
  user_date = input("Enter a date (YYYY-MM-DD) or press Enter for today: ")

  if user_date:
    url = f"https://api.nasa.gov/planetary/apod?api_key={API_KEY}&date={user_date}"
  else:
    url = f"https://api.nasa.gov/planetary/apod?api_key={API_KEY}"

  try:
    response = requests.get(url)
    if response.status_code == 200:
      data = response.json()

      title = data.get("title")
      img_url = data.get("url")
      explanation = data.get("explanation")
      media_type = data.get("media_type")

      print(f"\n MISSION LOG: {user_date if user_date else "TODAY"}")
      print(f"TITLE:{title}\n")

      if media_type == "image":
        display(Image(url=img_url,width=600))
      else:
        print(f"It's a video! Watch it here: {img_url}")

      print(f"\nThe Science:\n{explanation}")

    elif response.status_code == 400:
      print("Note: NASA's APOD archive only goes back to June 16, 1995.")
      print("Try a date like '2021-12-25' (James Webb Telescope Launch)!")
      print("Mission Failed: check data format(YYYY-MM-DD)")
    else :
      print(f"Error {response.status_code}.Space time is unstable")

  except Exception as e:
    print(f"System Crash: {e}")

launch_time_machine()

```

```python
def download_space_image(image_url, filename):
  try:
    response = requests.get(image_url, stream=True)
    if response.status_code == 200:
      with open(filename, "wb") as file:
        for chunk in response.iter_content(1024):
            file.write(chunk)

      print(f"Success! Image saved as:"{filename}")
      else:
            print("Failed to donwload")

  except Exception as e:
            print(f"Download error: {e}")
```
